## Transforming data to star schema & creating different table as dataframes

```python
col_list=['distributor', 'customer_name', 'city', 'country', 'channel', 'sub_channel', 'product_name', 'product_class',
        'name_of_sales_rep', 'manager', 'sales_team']
dfs={}
for col in col_list:
    
    dfs[f"dim_{col}"]=pd.DataFrame(df[col].drop_duplicates()).reset_index(drop=True).reset_index()
    dfs[f"dim_{col}"].columns=[col+'_id',col]
    df=pd.merge(df,dfs[f"dim_{col}"],on=[col])
    df.drop(col,axis=1,inplace=True)
    
```

```python
dfs["dim_name_of_sales_rep"]=pd.merge(dfs["dim_name_of_sales_rep"],df[['name_of_sales_rep_id','manager_id','sales_team_id']].drop_duplicates(),on=['name_of_sales_rep_id'])
df.drop(['manager_id','sales_team_id'],axis=1,inplace=True)
```

```python
dfs["dim_product_name"]=pd.merge(dfs["dim_product_name"],df[['product_name_id','product_class_id','price']].drop_duplicates(),on=['product_name_id'])
df.drop(['product_class_id','price'],axis=1,inplace=True)
```

## Save Tarnsformed Output to excel 

```python
file_name = 'pharama_data_clean.xlsx'

# Create an ExcelWriter object and write DataFrames to separate sheets
with pd.ExcelWriter(file_name, engine='xlsxwriter') as writer:
    df.to_excel(writer, sheet_name='fact_pharma', index=False)
    for file in dfs.keys():
        dfs[file].to_excel(writer, sheet_name=file, index=False)
```
